# nltk.sentiment

Get the list of positive words and negative words using nltk.

https://www.nltk.org/api/nltk.sentiment.html#module-nltk.sentiment

```python
sample = data.head(10)
```

```python
sid = SentimentIntensityAnalyzer()

pos_word_list=[]
neu_word_list=[]
neg_word_list=[]

neg = sample['filtered'][5] #this review is negative

for word in word_tokenize(neg):
    if (sid.polarity_scores(word)['compound']) >= 0.3 :
        pos_word_list.append(word)
    elif (sid.polarity_scores(word)['compound']) >= -0.3 :
        neg_word_list.append(word)
    else:
        neu_word_list.append(word)                

print('Positive :',pos_word_list)        
print('Neutral :',neu_word_list)    
print('Negative :',neg_word_list)    
```

# TextBlob

Get the list of positive words and negative words using TextBlob.

library : https://textblob.readthedocs.io/en/dev/

NaiveBayesAnalyzer in textblob library is the pre-trained model which outputs Sentiment(classification, p_pos, p_neg). However, it is not suitable for a token but for sentences. Instead, I used the polarity of each token and labeled it as below. 

- polarity > 0 : pos

- polarity < 0 : neg

```python
from textblob.sentiments import NaiveBayesAnalyzer

sample = data.head(10)

text = sample['filtered'][0]

sent          = TextBlob(text, analyzer = NaiveBayesAnalyzer())
classification= sent.sentiment.classification
positive      = sent.sentiment.p_pos
negative      = sent.sentiment.p_neg

print(classification, positive, negative)
```

```python

blob = TextBlob(neg)

textblob_pos = []
textblob_neg = []

for word in word_tokenize(neg):
    if (TextBlob(word).sentiment[0] > 0) :
        textblob_pos.append(word)
      
    elif (TextBlob(word).sentiment[0] < 0) :
        textblob_neg.append(word)
    else :
        pass

print('Positive :', textblob_pos)            
print('Negative :', textblob_neg)    
```

# Adding more feature variables

We would like to add the number of positive words and negative words per review as our independent variables.
For now, we used textblob library.

```python
# go through reviews and count the number of positive words and negative words per review.

len(data) #13250


num_pos = []
num_neg = []

# 'filtered' is already without stopwords and special characters
# but filter it once more
filtered = data['filtered'].map(lambda x: re.sub(r'\W+', ' ', str(x)))

for review in filtered :
    pos_counter = 0
    neg_counter = 0
    for word in word_tokenize(review):
        if(TextBlob(word).sentiment[0] > 0) :
            pos_counter = pos_counter+1
            #print(TextBlob(word).sentiment[0])
            
        elif (TextBlob(word).sentiment[0] < 0) :
            neg_counter = neg_counter+1
            #print(TextBlob(word).sentiment[0])
        
        else :
            pass
        
    num_pos.append(pos_counter)
    num_neg.append(neg_counter)
```

# Treating double negatives

i.e) legit phone not bad aspects : rating 5

Find lexicon for positive words and negative words <- tried but could not find prebuilt lexicon for double negatives.
We decided to create a list of double negative phrases from our dataset. Using bigrams, we will sift double negatives and create a list in order to nullify the double negatives and increase a positive word occurence counter.

1. Using bigrams, create a list of double negatives in our dataset.
2. sift the double negatives
3. go through the list from ratings 4 and 5 and retrieve double nagated words
4. and nullify that negation or give positive score ???

** I am going to work with filtered review. include '!' ? or not  ? 
problem for filtered : because it removed all the stop words, it removed all the hasn't, doesn't, does not .. etc .


```python
import re
from nltk.stem import WordNetLemmatizer
#from nltk.util import ngrams

sample = data.head(30)
#print(sample)
type(sample)

#remove non ascii characters and special characters
filtered = data['review'].str.replace(r'[^\x00-\x7f]', '')
filtered.replace(to_replace=r'[?|$|.|,|!|;|:|(|)|-]', value='',inplace=True, regex=True) 
```

```python
#Lemmatization before sifting double negatives

lemmatizer=WordNetLemmatizer()
tok = filtered.apply(lambda x : word_tokenize(x))
lem = tok.apply(lambda x : [lemmatizer.lemmatize(_) for _ in x])
```

```python
len(lem) #13250
lem_rat = pd.concat([lem, data['rating']], axis=1)
```

```python
lem_rat #lemmatized review and its star rating
```

```python
#list for bigrams of all the reviews
# also added rating next to each bigram
bigrams =[]
for i in range(len(lem_rat['review'])) :
    for ngram in (ngrams(lem_rat['review'][i], 2)) :
        bigrams.append(' '.join(str(i) for i in ngram))
        bigrams.append(lem_rat['rating'][i])

#bigrams #it has rating next to each bigrams. we'll use this rating later
```

```python
#list for possible double negatives
double_negatives = []

for i in range(0, len(bigrams), 2):
    tokens = word_tokenize(bigrams[i])
    #if both words are negative
    if(("no" in tokens[0]) & (TextBlob(tokens[1]).sentiment[0]<0)):
        double_negatives.append(bigrams[i])
        double_negatives.append(bigrams[i+1])
```

```python
double_negatives
```

```python
# let's separate the list and make it as a dataframe

d_neg = []
d_rat = []

i = 0
while(i < len(double_negatives)):
    d_neg.append(double_negatives[i])
    i=i+1
    d_rat.append(double_negatives[i])
    i=i+1

print(len(d_neg)) #201
print(len(d_rat)) #201

# list to a dataframe
d_neg_rat = pd.DataFrame({'negative bigrams' : d_neg, 'associated rating' : d_rat})
```

```python
d_neg_rat

pd.pivot_table(d_neg_rat, columns='associated rating', aggfunc='size')
```
